```python
user_books = books.pivot(index='Unnamed: 0', columns='Book.Title', values='Book.Rating').reset_index(drop=True)
user_books
```

```python
user_books.fillna(0,inplace=True)
user_books
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(user_books.values, metric='cosine')
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
user_sim_df.iloc[0:10, 0:10]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:10, 0:10]
```
