```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder#for train test splitting
from sklearn.model_selection import train_test_split#for decision tree object
from sklearn.tree import DecisionTreeClassifier#for checking testing results
from sklearn.metrics import classification_report, confusion_matrix#for visualizing tree 
from sklearn.tree import plot_tree
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
```

```python
comapny_data["ShelveLoc"]=le.fit_transform(comapny_data["ShelveLoc"])
comapny_data["Urban"]=le.fit_transform(comapny_data["Urban"])
comapny_data["US"]=le.fit_transform(comapny_data["US"])
```

```python
comapny_data=comapny_data.assign(Sale=pd.cut(comapny_data['Sales'], 
                               bins=[ 0, 4, 9,15], 
                               labels=['Low', 'Medium', 'High']))
```

```python
comapny_data.head(50)
```

```python
# let's plot pair plot to visualise the attributes all at once
sns.pairplot(data=comapny_data, hue = "Sale")
```

```python
target = pd.DataFrame.astype(comapny_data['Sale'], dtype="object")
df1 = comapny_data.copy()
df1 = df1.drop('Sale', axis =1)
```

```python
# Defining the attributes
X = df1
```

```python
target = target.fillna('').apply(str)
target
```

```python
#label encoding
target = le.fit_transform(target)
target
```

```python
y = target
```

```python
# Splitting the data - 80:20 ratio
X_train, X_test, y_train, y_test = train_test_split(X , y, test_size = 0.2, random_state =25)
print("Training split input- ", X_train.shape)
print("Testing split input- ", X_test.shape)
```

```python
# Defining the decision tree algorithm
dtree=DecisionTreeClassifier()
dtree.fit(X_train,y_train)
print('Decision Tree Classifier Created')
```

```python
# Predicting the values of test data
y_pred = dtree.predict(X_test)
```

```python
#2 Way table to understand the correct and wrong  predictions

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5,5))
sns.heatmap(data=cm,linewidths=.5, annot=True,square = True,  cmap = 'Blues')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
all_sample_title = 'Accuracy Score: {0}'.format(dtree.score(X_test, y_test))
plt.title(all_sample_title, size = 14)
```

```python
#PLot the decision tree
from sklearn import tree
tree.plot_tree(dtree);
```

```python
#Building Decision Tree Classifier using Entropy Criteria

model = DecisionTreeClassifier(criterion = 'entropy',max_depth=3)
model.fit(X_train,y_train)
```

```python
# Predicting the values of test data
y_pred1 = model.predict(X_test)
```

```python
#2 Way table to understand the correct and wrong  predictions

cm = confusion_matrix(y_test, y_pred1)
plt.figure(figsize=(5,5))
sns.heatmap(data=cm,linewidths=.5, annot=True,square = True,  cmap = 'Blues')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
all_sample_title = 'Accuracy Score: {0}'.format(model.score(X_test, y_test))
plt.title(all_sample_title, size = 14)
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```
